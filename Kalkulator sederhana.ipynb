```python
print("Select an operation to perform")
print("1. ADD")
print("2. SUBTRACT")
print("3. MULTIPLY")
print("4. DIVIDE")

operation = input()

if operation == "1":
  num1 = input ("Enter first Number: ")
  num2 = input ("Enter second Number: ")
  print("The sum is" + str(int(num1) + int(num2)) )
  #code for add
elif operation == "2":
  num1 = input ("Enter first Number: ")
  num2 = input ("Enter second Number: ")
  print("The sum is" + str(int(num1) - int(num2)) )
  #code for substract
elif operation == "3":
  num1 = input ("Enter first Number: ")
  num2 = input ("Enter second Number: ")
  print("The sum is" + str(int(num1) * int(num2)) )
  #code for  division
elif operation  == "4":
  num1 = input ("Enter first Number: ")
  num2 = input ("Enter second Number: ")
  print("The sum is" + str(int(num1) / int(num2)) )
  #code for multiply
else:
  print("Invalid Entry")
```
